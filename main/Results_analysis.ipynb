```python
import pandas as pd

# Load the datasets
combined_summary = pd.read_csv('../Results/Combined_summary.csv')
godel_summary = pd.read_csv('../Results/Godel_summary.csv')
product_summary = pd.read_csv('../Results/Product_summary.csv')
lukasiewic_summary = pd.read_csv('../Results/Lukasiewic_summary.csv')
F2_summary = pd.read_csv('../Results/F2_summary.csv')
F4_summary = pd.read_csv('../Results/F4_summary.csv')
Root_node_summary = pd.read_csv('../Results/root_node_summary.csv')
Root_node_max_summary = pd.read_csv('../Results/root_node_max_summary.csv')


# Define a function to analyze each dataset
def analyze_dataset(df):
    # Find the row with the highest F1 score
    highest_f1_row = df.loc[df['F1 score'].idxmax()]
    return highest_f1_row

# Perform the analysis for each dataset
combined_analysis = analyze_dataset(combined_summary)
godel_analysis = analyze_dataset(godel_summary)
product_analysis = analyze_dataset(product_summary)
lukasiewic_analysis = analyze_dataset(lukasiewic_summary)
F2_analysis = analyze_dataset(F2_summary)
F4_analysis = analyze_dataset(F4_summary)
Root_node_analysis = analyze_dataset(Root_node_summary)
Root_node_max_analysis = analyze_dataset(Root_node_max_summary)

# Create a DataFrame to store the results
results_df = pd.DataFrame(columns=['Dataset', 'Best Expression', 'F1 Score', 'Complexity', 'Precision', 'Recall', 'Accuracy'])

# Collect the results in a DataFrame
datasets = ['Combined_summary', 'Godel_summary', 'Product_summary', 'Lukasiewic_summary', 'F2_summary', 'F4_summary', 'Root_node_summary', 'Root_node_max_summary']
analyses = [combined_analysis, godel_analysis, product_analysis, lukasiewic_analysis, F2_analysis, F4_analysis, Root_node_analysis, Root_node_max_analysis]

for dataset, analysis in zip(datasets, analyses):
    results_df = results_df.append({
        'Dataset': dataset,
        'Best Expression': analysis['Evaluated expression'],
        'F1 Score': analysis['F1 score'],
        'Complexity': analysis['complexity'],
        'Precision': analysis['precision'],
        'Recall': analysis['recall'],
        'Accuracy': analysis['accuracy']
    }, ignore_index=True)

results_df
```

```python
import pandas as pd

# Load the datasets
combined_summary = pd.read_csv('../Results/Combined_summary.csv')
godel_summary = pd.read_csv('../Results/Godel_summary.csv')
product_summary = pd.read_csv('../Results/Product_summary.csv')
lukasiewic_summary = pd.read_csv('../Results/Lukasiewic_summary.csv')
F2_summary = pd.read_csv('../Results/F2_summary.csv')
F4_summary = pd.read_csv('../Results/F4_summary.csv')
Root_node_summary = pd.read_csv('../Results/root_node_summary.csv')
Root_node_max_summary = pd.read_csv('../Results/root_node_max_summary.csv')

# Define a function to analyze each dataset
def analyze_dataset(df):
    # Group by seed and find the highest F1 score per seed
    grouped_by_seed = df.groupby('seed').apply(lambda x: x.loc[x['F1 score'].idxmax()])
    
    # Calculate the average and standard deviation of the highest F1 scores per seed
    average_f1_per_seed = grouped_by_seed['F1 score'].mean()
    average_precision_per_seed = grouped_by_seed['precision'].mean()
    average_recall_per_seed = grouped_by_seed['recall'].mean()
    average_accuracy_per_seed = grouped_by_seed['accuracy'].mean()
    average_complexity_per_seed = grouped_by_seed['complexity'].mean()
    
    return average_f1_per_seed, average_recall_per_seed, average_precision_per_seed, average_accuracy_per_seed, average_complexity_per_seed

# Perform the analysis for each dataset
combined_analysis = analyze_dataset(combined_summary)
godel_analysis = analyze_dataset(godel_summary)
product_analysis = analyze_dataset(product_summary)
lukasiewic_analysis = analyze_dataset(lukasiewic_summary)
F2_analysis = analyze_dataset(F2_summary)
F4_analysis = analyze_dataset(F4_summary)
Root_node_analysis = analyze_dataset(Root_node_summary)
Root_node_max_analysis = analyze_dataset(Root_node_max_summary)

# Create a DataFrame to store the results
results_df = pd.DataFrame(columns=[
    'Dataset', 'Average F1 Score', 'Average Precision', 'Average Recall', 'Average Accuracy', 'Average Complexity'
])

# Collect the results in a DataFrame
datasets = ['Combined_summary', 'Godel_summary', 'Product_summary', 'Lukasiewic_summary', 'F2_summary', 'F4_summary', 'Root_node_summary', 'Root_node_max_summary']
analyses = [combined_analysis, godel_analysis, product_analysis, lukasiewic_analysis, F2_analysis, F4_analysis, Root_node_analysis, Root_node_max_analysis]

for dataset, analysis in zip(datasets, analyses):
    average_f1_per_seed, average_precision_per_seed, average_recall_per_seed, average_accuracy_per_seed, average_complexity_per_seed = analysis
    results_df = results_df.append({
        'Dataset': dataset,
        'Average F1 Score': average_f1_per_seed,
        'Average Precision': average_precision_per_seed,
        'Average Recall': average_recall_per_seed,
        'Average Accuracy': average_accuracy_per_seed,
        'Average Complexity': average_complexity_per_seed
    }, ignore_index=True)

results_df
```
